```python
def generate_embeddings(text, model_name="en_core_web_lg"):
    nlp = spacy.load(model_name)
    doc = nlp(text)
    return doc.vector.tolist()

text = 'Я люблю вечірній квартал'
embedding = generate_embeddings(translate(text))
print("Generated Embedding:")
print(embedding)
```

```python
from transformers import BertTokenizer, BertModel
import torch

def generate_bert_embeddings(text, model_name="bert-base-uncased"):
    tokenizer = BertTokenizer.from_pretrained(model_name)
    model = BertModel.from_pretrained(model_name)

    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)

    with torch.no_grad():
        outputs = model(**inputs)

    embeddings = outputs.last_hidden_state.squeeze(0)

    aggregated_embedding = embeddings.mean(dim=0)

    embedding_list = aggregated_embedding.tolist()

    return embedding_list
```

```python
text = "Я люблю вечірній квартал"
embedding = generate_bert_embeddings(translate(text))
print("Generated Embedding:")
print(embedding)
```

```python
from tqdm import tqdm
def process_row(row):
    text = row['desc']
    if pd.isna(text) or text.strip() == "":
        return row
    spacy_en_embedding = generate_embeddings(text, "en_core_web_lg")

    translated_text = translate(text)

    spacy_uk_embedding = generate_embeddings(translated_text, "uk_core_news_lg")


    row['spacy_en'] = spacy_en_embedding
    row['spacy_uk'] = spacy_uk_embedding


    return row

passat_first_500 = passat.head(500)

tqdm.pandas(desc="Processing first 500 rows")

passat_first_500 = passat_first_500.progress_apply(process_row, axis=1)
```
